```python

# Importing necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("dentalAppointments.csv")
demo = pd

# Display the first few rows of the dataset
df.head()
```

```python

# Check for missing values
df.isnull().sum()
```

```python

# Set style
sns.set_style("whitegrid")
```

```python
# 1. Age Distribution of Patients
plt.figure(figsize=(8,5))
sns.histplot(df["Age"], bins=10, kde=True, color="blue")
plt.title("Age Distribution of Patients")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
# 2. Gender Distribution
plt.figure(figsize=(6,5))
sns.countplot(x="Gender", data=df, palette="pastel")
plt.title("Gender Distribution of Patients")
plt.show()
```

```python
# 3. Treatment Type Distribution
plt.figure(figsize=(10,5))
sns.countplot(y="Treatment Type", data=df, order=df["Treatment Type"].value_counts().index, palette="viridis")
plt.title("Distribution of Treatment Types")
plt.xlabel("Count")
plt.ylabel("Treatment Type")
plt.show()
```

```python
# 4. Appointment Status Count
plt.figure(figsize=(7,5))
sns.countplot(x="Status", data=df, palette="coolwarm")
plt.title("Appointment Status Distribution")
plt.show()
```

```python
# 5. Cost Distribution
plt.figure(figsize=(8,5))
sns.histplot(df["Cost ($)"], bins=10, kde=True, color="green")
plt.title("Cost Distribution of Treatments")
plt.xlabel("Cost ($)")
plt.ylabel("Count")
plt.show()
```

```python
# 6. Average Cost per Treatment Type
plt.figure(figsize=(12,5))
sns.barplot(y="Treatment Type", x="Cost ($)", data=df, estimator=sum, ci=None, palette="magma")
plt.title("Average Cost per Treatment Type")
plt.xlabel("Total Cost ($)")
plt.ylabel("Treatment Type")
plt.show()
```

```python
# 7. Number of Appointments per Dentist
plt.figure(figsize=(10,5))
sns.countplot(y="Dentist", data=df, order=df["Dentist"].value_counts().index, palette="Set2")
plt.title("Number of Appointments per Dentist")
plt.xlabel("Count")
plt.ylabel("Dentist")
plt.show()
```

```python
# 8. Monthly Appointment Trend
df["Appointment Date"] = pd.to_datetime(df["Appointment Date"])
df["Month"] = df["Appointment Date"].dt.month_name()
plt.figure(figsize=(12,5))
sns.countplot(x="Month", data=df, order=pd.date_range(start="2024-01-01", periods=12, freq='M').strftime('%B'), palette="husl")
plt.title("Monthly Appointment Trend")
plt.xticks(rotation=45)
plt.show()
```

```python
# 9. Cost vs Age Scatter Plot
plt.figure(figsize=(8,5))
sns.scatterplot(x="Age", y="Cost ($)", data=df, color="red")
plt.title("Age vs Cost of Treatment")
plt.xlabel("Age")
plt.ylabel("Cost ($)")
plt.show()
```

```python
# 10. Dentist-wise Treatment Types Heatmap
pivot_table = df.pivot_table(index="Dentist", columns="Treatment Type", aggfunc="size", fill_value=0)
plt.figure(figsize=(12,6))
sns.heatmap(pivot_table, annot=True, cmap="coolwarm", fmt="d")
plt.title("Dentist-wise Treatment Type Frequency")
plt.xlabel("Treatment Type")
plt.ylabel("Dentist")
plt.show()
```
